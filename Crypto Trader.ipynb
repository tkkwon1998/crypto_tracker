```python
def getData(client, granularity, ticker, days):
    # lower bound for the date
    lowerBound = datetime.now() - timedelta(days=days)
    
    # number of data points per day
    splits = 24 / (granularity / 60 / 60)
    
    # number of days we can get data for (data segments)
    day_calls = 200 / splits
    
    end = datetime.now()
    start = end - timedelta(days=day_calls)
    data = []
    for _ in range(math.ceil(days // day_calls)):
        next = client.get_product_historic_rates(ticker, start=start, end=end, granularity=granularity)
        data.extend(next)
        start -= timedelta(days=day_calls)
        end -= timedelta(days=day_calls)
    try:
        data = pd.DataFrame(data, columns=["time", "low", "high", "open", "close", "volume"])
        data['ticker'] = ticker
        data['date'] = pd.to_datetime(data['time'],unit='s')
        return data
    
    except:
        return pd.DataFrame()
```
