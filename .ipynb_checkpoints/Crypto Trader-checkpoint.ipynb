```python
def getData(client, granularity, ticker, days):
    # lower bound for the date
    lowerBound = datetime.now() - timedelta(days=days)
    
    # number of data points per day
    splits = 24 / (granularity / 60 / 60)
    
    # number of days we can get data for (data segments)
    day_calls = 200 / splits
    
    end = datetime.now()
    start = end - timedelta(days=day_calls)
    data = []
    for _ in range(math.ceil(days // day_calls)):
        next = client.get_product_historic_rates(ticker, start=start, end=end, granularity=granularity)
        data.extend(next)
        start -= timedelta(days=day_calls)
        end -= timedelta(days=day_calls)
    try:
        data = pd.DataFrame(data, columns=["time", "low", "high", "open", "close", "volume"])
        data['ticker'] = ticker
        data['date'] = pd.to_datetime(data['time'],unit='s')
        return data
    
    except:
        return pd.DataFrame()

def getRank(data, ticker, lowerBound, skip):
    try:
        df = data[data['ticker'] == ticker].iloc[skip:lowerBound]
        df['percentChange'] = 100 * ((df['close'] - df['open']) / df['open'])
        reg = stats.linregress(df['time'].values, df['percentChange'].values)
        return (token, reg.rvalue, reg.slope, reg.rvalue ** 2 * reg.slope)
    except:
        return None
```

```python
data = pd.read_csv("coinbaseData.csv")
```

```python
holding_periods = list(range(4, 20, 4))
windows = list(range(40, 120, 12))
lags = list(range(20, 120, 12))
percent = [.25, .25, .2, .2, .1]
transactionFee = .01
cash = 35000
results = []
daysAgoToStartTesting = 30

iterations = len(holding_periods) * len(windows) * len(lags)
curr = 0
for holding_period in holding_periods:
    for window in windows:
        amounts = []
        skip = daysAgoToStartTesting * 4
        lowerDayBound = skip + window
        for lag in lags:
            print(curr/iterations)
            curr += 1
            
            openCloseIndex = skip - lag
            
            if openCloseIndex < 0:
                continue
            
            originalBTCPrice = data[data['ticker'] == "BTC-USD"].iloc[openCloseIndex]['open']
            wallet = {}
            for i in range(openCloseIndex // holding_period):
                ranks = []
                for token in tokens:
                    rank = getRank(data, token, lowerDayBound, skip)
                    if rank:
                        ranks.append(rank)

                ranks.sort(key=lambda x : x[3], reverse=True)

                amount = 0 if len(wallet) else cash / originalBTCPrice
                for coin in wallet:
                    amount += wallet[coin] * data[data['ticker'] == coin].iloc[openCloseIndex]['open']
                amounts.append(amount)

                # Get top 5 coins
                wallet = {}
                for i in range(5):
                    coin = ranks[i][0]
                    wallet[coin] = (amount * percent[i] * (1 - transactionFee)) / data[data['ticker'] == coin].iloc[openCloseIndex]['open']

                openCloseIndex -= holding_period
                lowerDayBound -= holding_period
                skip -= holding_period
            
            btcDiff = amounts[-1] - amounts[0]
            profit = data[data['ticker'] == "BTC-USD"].iloc[0]['open'] * amounts[-1] - cash
            portfolioSTD = np.std(amounts)
            sharp = ((amounts[-1] - amounts[0]) / amounts[0]) / portfolioSTD
            
            results.append((btcDiff, profit, portfolioSTD, sharp, amounts, holding_period, window, lag))
            
            plt.plot(list(range(len(amounts))), amounts)
            plt.xlabel("Time Periods since the start")
            plt.ylabel("BTC amount")
            plt.title(str((btcDiff, profit, portfolioSTD, sharp, holding_period, window, lag)))
            plt.show()
```
